```python
def egWtPerToSatEgVol(wt):
    egDensity=2260
    parDensity=900
    parPerSatEG=.977
    porDensity=1/250
    wormVol=100/porDensity
    wormParVol=wormVol*parPerSatEG
    egVol=wormVol*(1-parPerSatEG)
    sParVol=((egVol*egDensity)/wt-wormParVol*parDensity-egVol*egDensity)/parDensity
    return (egVol+wormParVol)/(egVol+sParVol+wormParVol)

def egWtPerToSParVol(wt):
    egDensity=2260
    parDensity=900
    parPerSatEG=.977
    porDensity=1/250
    wormVol=100/porDensity
    wormParVol=wormVol*parPerSatEG
    egVol=wormVol*(1-parPerSatEG)
    sParVol=((egVol*egDensity)/wt-wormParVol*parDensity-egVol*egDensity)/parDensity
    return sParVol/(egVol+sParVol+wormParVol)

i=0.01
egVolList=[]
sParVolList=[]
wtPerList=[]
while i<=0.056:
    print(round(i, 3), egWtPerToSatEgVol(i))
    egVolList.append(egWtPerToSatEgVol(i))
    sParVolList.append(egWtPerToSParVol(i))
    wtPerList.append(i)
    i+=0.001
```
